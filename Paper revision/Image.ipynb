```python
import numpy as np
import matplotlib.pyplot as plt

# Define image dimensions and pixel size
width = 100
height = 100
pixel_size = 10  # in microns

# Define mass range and number of peaks
mass_min = 500
mass_max = 10000
num_peaks = 10

# Define peak parameters (mass and intensity)
masses = np.linspace(mass_min, mass_max, num_peaks)
intensities = np.random.rand(num_peaks) * 100

# Create x and y coordinates for each pixel
x_coords = np.arange(width) * pixel_size
y_coords = np.arange(height) * pixel_size

# Create grid of x and y coordinates
xx, yy = np.meshgrid(x_coords, y_coords)

# Initialize image as zeros
image = np.zeros((height, width))

# Loop over each peak and add it to the image
for mass, intensity in zip(masses, intensities):
    # Generate Gaussian peak at random position
    x_peak = np.random.randint(0, width)
    y_peak = np.random.randint(0, height)
    sigma = mass / 1000  # width of peak depends on mass
    peak = intensity * np.exp(-((xx - x_peak) ** 2 + (yy - y_peak) ** 2) / (2 * sigma ** 2))
    # Add peak to image
    image += peak

# Display image
plt.imshow(image, cmap='gray')
plt.axis('off')
plt.show()
```
